```python
import json
```

```python
with open('location_plot.json') as json_file:
    data = json.load(json_file)
```

```python
data
```

```python
locationData = []
for value in data:
    point = value['multi_geo']
    for points in point:
        info = {}
        if 'lat' not in info:
            info['lat'] = []
        if 'lng' not in info:
            info['lnf'] = []
        info['lat'] = points['geocode']['lat']
        info['lng'] = points['geocode']['lng']   
        locationData.append(info)

        
```

```python
for value in locationData:
    del value['lnf']
```

```python
with open('location_data.json','w') as output_file:
    output_file.write(str(locationData))
```
